###Conclusion

If I was working for a financial institution evaluating machine learning algorithms for optical character recognition and testing this MNIST dataset, I would definitely use the TensorFlow machine learning library to analyze the dataset and produce as accurate as possible results in predicting the characters. With TensorFlow, you are allowed to set parameters freely and find whichever fits the best for your model. 

Something I found when working with the dataset is that familiarization with the model is better than how many nodes you set throughout each layer and hidden layer in your neural network. I realized this while rerunning some of the models and seeing the prediction accuracy scores hit new highs every time. I believe the same results would be possible with higher epochs set, so that the model can learn more about the data and better fit the model for more accurate predictions. However, the number of nodes do have a significant amount to do with how good the model fits to the data. Too many nodes could lead to overfitting. The best model I would choose out of the ones I have run would be 2 layers with 50 nodes per layer and 150 epochs so it has enough time to best fit the data after learning as much as it can from it.
