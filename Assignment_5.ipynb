###KMeans Pros and Cons

```
import multiprocessing

print(f"Total CPU Count: {multiprocessing.cpu_count()}")
```

```
from multiprocessing import Pool
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
import time

def do_kmeans(n_samples):
    """KMeans clustering on generated data"""

    X,_ = make_blobs(n_samples, centers=3, n_features=10,
                random_state=0)
    kmeans = KMeans(n_clusters=3)
    t0 = time.time()
    kmeans.fit(X)
    print(f"KMeans cluster fit in {time.time()-t0}")

def run_parallel():
    """Run Everything"""

    count = 25
    processes = 13
    t0 = time.time()
    with Pool(processes) as p:
        p.map(do_kmeans, [100000 for x in range(count)])
    print(f"Performed {count} KMeans in total time: {time.time()-t0} using processes total: {processes}")

run_parallel()
```

This segment of code runs the KMeans clustering on a dataset called make_blobs, which seems to be a random sample generator. It runs the KMeans clustering algorithm with threedifferent clusters, and hence cetroids, to group them and find the SSEs accordingly. It also times how long it would take to run the algorithm for each sample run, of which it does 25 different runs. It pools the sample sizes into 100,000; hence why this process is longer than the next. Each process takes longer in this example, however they are done at once rather than individually, reducing the overall time used.

```
from sklearn.datasets.samples_generator import make_blobs
from sklearn.cluster import KMeans
import time

def do_kmeans():
    """KMeans clustering on generated data"""

    X,_ = make_blobs(n_samples=100000, centers=3, n_features=10,
                random_state=0)
    kmeans = KMeans(n_clusters=3)
    t0 = time.time()
    kmeans.fit(X)
    print(f"KMeans cluster fit in {time.time()-t0}")

def run_serial():
    """Run Everything"""

    count = 25
    processes = 0
    t0 = time.time()
    for _ in range(count):
        do_kmeans()
    print(f"Performed {count} KMeans in total time: {time.time()-t0} using processes total: {processes}")
run_serial()
```

This segment of code runs the KMeans clustering on a dataset called make_blobs, which seems to be a random sample generator. It runs the KMeans clustering algorithm with threedifferent clusters, and hence cetroids, to group them and find the SSEs accordingly. It also times how long it would take to run the algorithm for each sample run, of which it does 25 different runs. This way uses less time per process, but does them one by one, hence causing a longer overall runtime.

Pros and Cons

The cons of KMeans clustering could be determining which amount of clusters is the right one, as getting the wrong one could be detrimental to perfomance. When data is so close together, it may not be able to determine which belongs in which cluster, and hence may lead to inaccurate results in the end. However, tht is a risk in lots of machine learning algorithms, so may not apply directly to this. Another con could be the time it takes to run each individual one, which could be necessary to find the right number of clusters you should pick.


The pros of KMeans clustering are that it is relatively simple to model, and could be used on large datasets with quite some ease. The clusters could also be various shapes to better fit the data, leading to more efficiency and accuracy than some other models.
